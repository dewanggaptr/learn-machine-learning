```python
import pandas as pd

data = pd.read_excel('data/dataset_quiz.xlsx')

data.head(100)
```

```python
# melakukan backup data ke var dtest
dtest = data.copy()
```

```python
# menghapus kolom id_anggota5 karena bernilai null
dtest = dtest.drop(dtest.columns[[5]], axis=1)
```

```python
# Cek kolom null
dtest.isnull().sum()
```

```python
# mengganti value NA(null) ke 0
dtest =dtest.fillna(0)
```

```python
#Cek kembali apakah masih terdapat kolom null
dtest.isnull().sum()
```

```python
# cek data duplikasi
dtest.duplicated().sum()
```

```python
# Cek informasi sata
print(dtest.info())
```

```python
dtb = dtest.copy() # backup data

#melakukan seleksi fitur seluruh kolom kecual kolom id
dtb = dtb.drop(dtb.iloc[:,:5], axis=1)
dtb = dtb.drop(dtb.iloc[:,2:3], axis=1)

dtb.head(100)
```

```python
# melakukan labelling pada kolom
from sklearn.preprocessing import LabelEncoder

labelEn = LabelEncoder()
for col in dtb.columns:
    dtb[col] = labelEn.fit_transform(dtb[col])

display(dtb.head())
```

```python
#melakukan seleksi fitur
X = dtb.iloc[:, 1:].values
y = dtb['keterlibatan_vltr'].values

X.shape
```

```python
# melakukan split dan training data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
# metode klasifikasi dengan Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# menyesuaikan dt ke set training
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

# Memprediksi label set test
y_pred_dt = dt.predict(X_test)

# Menghitung set accuracy
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Test set accuracy: {acc_dt}")
```

```python
# metode klasifikasi dengan RandomForest
from sklearn.ensemble import RandomForestClassifier

# menyesuaikan rf ke set training
rf = RandomForestClassifier(n_estimators=30, random_state=42)
rf.fit(X_train, y_train)

# Memprediksi label set test
y_pred_rf = rf.predict(X_test)

# Menghitung set accuracy
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Test set accuracy: {:.2f}".format(acc_rf))
print(f"Test set accuracy: {acc_rf}")
```

```python
from sklearn.model_selection import RepeatedKFold

# inisiasi repated k-fold
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

```python
# Gridsearch menggunakan DecisionTree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# inisiasi model
dte = DecisionTreeClassifier()

# Melakukan pedfinisian hyperparameter yang akan digunakan
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11)),
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = GridSearchCV(dte, param_grid=params, cv=cv, n_jobs=-1)

%timeit grid.fit(X_train, y_train)

# melakukan evaluasi dengan score
score = grid.score(X_train, y_train)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

dtr = DecisionTreeClassifier()

param = {
    'criterion': ['gini', 'entropy'],
    'max_depth': list(range(5,11))
}

# inisiasi randoms berdasarkan nilai repeated k-fold dan hyperparameter
randoms = RandomizedSearchCV(dtr, param_distributions=param, cv=cv, n_jobs=-1)

%timeit randoms.fit(X_train, y_train)

# Evaluasi dengan score
score = randoms.score(X_train, y_train)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {randoms.best_params_}')
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
 
mnb = MultinomialNB()

mnb.fit(X_train, y_train)

# melakukan prediksi data train
y_pred_train = mnb.predict(X_train)
acc_train = accuracy_score(y_train, y_pred_train)

# melakukan prediksi data test
y_pred_test = mnb.predict(X_test)
acc_test = accuracy_score(y_test, y_pred_test)

print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc = SVC(kernel="linear")

svc.fit(X_train, y_train)

# melakukan prediksi data train
y_pred_trains = svc.predict(X_train)
acc_trains = accuracy_score(y_train, y_pred_trains)

# melakukan prediksi data test
y_pred_tests = svc.predict(X_test)
acc_tests = accuracy_score(y_test, y_pred_tests)

print(f'Hasil akurasi data train: {acc_trains}')
print(f'Hasil akurasi data test: {acc_tests}')
```
