```python
#import data
data = pd.read_csv("data/insurance.csv")
data.head()
```

```python
#data mapping
label_sex = {
    'male' : 1, 
    'female': 0
}

data['sex'] = data['sex'].map(label_sex)
data.head()
```

```python
#data mapping
label_region = {
    'northwest': 0,
    'northeast': 1,
    'southwest' : 2,
    'southeast' : 3
}

data['region'] = data['region'].map(label_region)
data.head(10)
```

```python
#data mapping
label_smoker = {
    'yes' : 1,
    'no': 0
}

data['smoker'] = data['smoker'].map(label_smoker)
data.head(10)
```

```python
#cek multicolinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(data):

    # menghitung VIF
    vif = pd.DataFrame()
    vif["variables"] = data.columns
    vif["VIF"] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]

    return(vif)

```

```python
data_baru = data.iloc[:,:-1]
calc_vif(data_baru)
```

```python
data_ku = data.copy()
data_ku = data_ku.drop(columns=['bmi'])
data_ku.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
x_m = data_ku.iloc[:, 0:4].values
y_m = data_ku.iloc[:, -1].values

x_m_train, x_m_test, y_m_train, y_m_test = train_test_split(x_m, y_m, test_size = 0.2, random_state = 30)

y_m = y_m.reshape(len(y_m), 1)

l_reg = LinearRegression()

l_reg.fit(x_m_train, y_m_train)

y_m_pred = l_reg.predict(x_m_test)

data_ku = pd.DataFrame({'Actual': y_m_test.flatten(), 'Predicted': y_m_pred.flatten()})

print('R2 Score:', r2_score(y_m_test, y_m_pred))
```
